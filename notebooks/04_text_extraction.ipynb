```python
import easyocr

print("EasyOCR loaded successfully!")
```

```python
reader = easyocr.Reader(['en'])

print("OCR reader created successfully!")
```

```python
results = reader.readtext("../invoices/sample.jpg")

print("Text extraction completed!")
print("Number of detected text regions:", len(results))
```

```python
for item in results:
    text = item[1]
    confidence = item[2]
    print(f"{text}  | confidence: {confidence:.2f}")
```

```python
results_pdf = reader.readtext("../outputs/page1_high_quality.png")

for item in results_pdf:
    text = item[1]
    confidence = item[2]
    print(f"{text} | confidence: {confidence:.2f}")
```
