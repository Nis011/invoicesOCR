```python
import fitz

pdf = fitz.open("../invoices/sample.pdf")

print("PDF opened successfully!")
print("Number of pages:", len(pdf))
```

```python
page = pdf[0]

pix = page.get_pixmap()

pix.save("../outputs/page1.png")

print("Page converted to image successfully!")
```

```python
page = pdf[0]

zoom = 3
matrix = fitz.Matrix(zoom, zoom)

pix = page.get_pixmap(matrix=matrix)
pix.save("../outputs/page1_high_quality.png")

print("High-quality page image saved successfully!")
```
